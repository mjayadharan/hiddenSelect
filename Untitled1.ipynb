```julia
using DifferentialEquations, Optim, Random, LinearAlgebra, Plots, Distributions

# ---------------------------------------------------------------
# 1) Define a 2D ODE system: Lotka-Volterra (predator-prey)
# ---------------------------------------------------------------
function lotka_volterra!(du, u, p, t)
    α, β, γ, δ = p
    x, z = u
    du[1] = α * x - β * x * z
    du[2] = δ * x * z - γ * z
end

# Jacobian of the Lotka-Volterra system
function jac_lotka_volterra!(J, u, p, t)
    α, β, γ, δ = p
    x, z = u
    J[1, 1] = α - β * z
    J[1, 2] = -β * x
    J[2, 1] = δ * z
    J[2, 2] = δ * x - γ
end

# ---------------------------------------------------------------
# 2) Generate synthetic data using a standard solver
# ---------------------------------------------------------------
function generate_data(α, β, γ, δ, t_data, y0; noise_std=0.01)
    p = (α, β, γ, δ)
    prob = ODEProblem(lotka_volterra!, y0, (t_data[1], t_data[end]), p)
    sol = solve(prob, Tsit5(), saveat=t_data, reltol=1e-10, abstol=1e-10)
    y_clean = Array(sol)
    y_noisy = y_clean .+ noise_std * randn(size(y_clean))
    return y_noisy
end

# ---------------------------------------------------------------
# 3) Fixed-step Euler integrator
# ---------------------------------------------------------------
function euler_integration(fun, t0, tf, y0, n_points, params)
    dt = (tf - t0) / (n_points - 1)
    t_sol = LinRange(t0, tf, n_points)
    y_sol = zeros(length(y0), n_points)
    y_sol[:, 1] .= y0

    for i in 2:n_points
        dydt = similar(y_sol[:, i-1])
        fun(dydt, y_sol[:, i-1], params, t_sol[i-1])
        y_sol[:, i] .= y_sol[:, i-1] + dt * dydt
    end
    return t_sol, y_sol'
end

function backward_euler_integration(fun, jac, t0, tf, y0, n_points, params; max_iter=50, tol=1e-8)
    dt = (tf - t0) / (n_points - 1)
    t_sol = LinRange(t0, tf, n_points)
    y_sol = zeros(length(y0), n_points)
    y_sol[:, 1] .= y0

    for n in 2:n_points
        t_next = t_sol[n]
        y_new = copy(y_sol[:, n-1])
        
        for _ in 1:max_iter
            dydt = similar(y_new)
            fun(dydt, y_new, params, t_next)
            G = y_new - y_sol[:, n-1] - dt * dydt
            
            if norm(G) < tol
                break
            end
            
            J = zeros(2, 2)
            jac(J, y_new, params, t_next)
            J_G = I - dt * J
            
            delta_y = -J_G \ G
            y_new += delta_y
            
            if norm(delta_y) < tol
                break
            end
        end
        y_sol[:, n] .= y_new
    end

    return t_sol, y_sol'
end

# ---------------------------------------------------------------
# 4) Cost function for optimization
# ---------------------------------------------------------------
function cost_function(params, t_data, data, y0, n_euler_points)
    α, β, γ, δ = params
    t0, tf = t_data[1], t_data[end]
    t_euler, y_euler = backward_euler_integration(lotka_volterra!, jac_lotka_volterra!,
                                                  t0, tf, y0, n_euler_points, (α, β, γ, δ))

    if any(isnan, y_euler) || any(isinf, y_euler)
        return Inf  # Penalize bad results
    end
    
    residuals = []
    for t_i in t_data
        idx = argmin(abs.(t_euler .- t_i))
        push!(residuals, y_euler[idx, :] .- data[:, findfirst(t_data .== t_i)])
    end

    residual_sum = sum(norm.(residuals).^2)
    return isfinite(residual_sum) ? residual_sum : Inf
end

# ---------------------------------------------------------------
# 5) Main demonstration
# ---------------------------------------------------------------
Random.seed!(0)
α_true, β_true, γ_true, δ_true = 1.0, 0.5, 0.8, 0.3
n_data = 50
t_data = LinRange(0, 19, n_data)
y0 = [1.0, 0.5]

data_noisy = generate_data(α_true, β_true, γ_true, δ_true, t_data, y0; noise_std=0)

euler_points_list = [200*j for j in 1:9]
fitted_params = []

initial_guess = rand(Uniform(2.5, 3.5), 4)
# initial_guess = [0.9,0.6,0.7,0.2]

for n_euler in euler_points_list
    
    # result = optimize(p -> cost_function(p, t_data, data_noisy, y0, n_euler), initial_guess, BFGS())
    result = optimize(p -> cost_function(p, t_data, data_noisy, y0, n_euler), initial_guess, NelderMead())
    
    push!(fitted_params, Optim.minimizer(result))
end

# Print results
println("True parameters:")
println("  α=$α_true, β=$β_true, γ=$γ_true, δ=$δ_true\n")

println("n_euler_points -> (α_est, β_est, γ_est, δ_est)")
for (n_euler, params) in zip(euler_points_list, fitted_params)
    println("$(n_euler) -> (α=$(params[1]), β=$(params[2]), γ=$(params[3]), δ=$(params[4]))")
end

# ---------------------------------------------------------------
# 6) Plot the results
# ---------------------------------------------------------------
p = plot(layout=(1, length(euler_points_list)), size=(18 * 80, 4 * 80), link=:y)

for (i, (n_euler, params)) in enumerate(zip(euler_points_list, fitted_params))
    α, β, γ, δ = params
    t_sim = LinRange(0, 19, 100)
    y_sim = generate_data(α, β, γ, δ, t_sim, y0, noise_std=0)
    
    plot!(p[i], t_data, data_noisy[1, :], seriestype=:scatter, label="Data x")
    plot!(p[i], t_data, data_noisy[2, :], seriestype=:scatter, label="Data z")
    plot!(p[i], t_sim, y_sim[1, :], label="Euler x")
    plot!(p[i], t_sim, y_sim[2, :], label="Euler z")
    title!(p[i], "Euler points = $n_euler")
end

display(p)
```
