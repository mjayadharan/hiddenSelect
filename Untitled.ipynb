```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import minimize

# ---------------------------------------------------------------
# 1) Define a 2D ODE system: Lotka-Volterra (predator-prey)
# ---------------------------------------------------------------
def lotka_volterra(t, y, alpha, beta, gamma, delta):
    """
    Lotka-Volterra system (predator-prey):
      dx/dt = alpha*x - beta*x*z
      dz/dt = delta*x*z - gamma*z

    y = [x, z]:
      x = prey population
      z = predator population
    """
    x, z = y
    dxdt = alpha * x - beta * x * z
    dzdt = delta * x * z - gamma * z
    return [dxdt, dzdt]

import numpy as np

def jac_lotka_volterra(t, y, alpha, beta, gamma, delta):
    """
    Jacobian of f(t, [x, z]) for the Lotka-Volterra system
    with respect to y = [x, z].
    
    f1 = alpha*x - beta*x*z
    f2 = delta*x*z - gamma*z
    
    Returns a 2x2 NumPy array:
      [[df1/dx, df1/dz],
       [df2/dx, df2/dz]]
    """
    x, z = y
    dfdx = np.array([
        [alpha - beta*z,   -beta*x   ],
        [delta*z,          delta*x - gamma]
    ])
    return dfdx




# ---------------------------------------------------------------
# 2) Generate synthetic data using a standard solver
#    at 20 data points
# ---------------------------------------------------------------
def generate_data(alpha_true, beta_true, gamma_true, delta_true,
                  t_data, y0, noise_std=0.01):
    """
    Generate 'true' data by solving with a standard ODE solver (solve_ivp)
    using tight tolerances. Evaluate solution exactly at t_data.
    Add optional Gaussian noise.
    """
    def rhs(t, y):
        return lotka_volterra(t, y, alpha_true, beta_true, gamma_true, delta_true)

    sol = solve_ivp(rhs, [t_data[0], t_data[-1]], y0,
                    t_eval=t_data, rtol=1e-10, atol=1e-10)
    y_clean = sol.y.T  # shape (len(t_data), 2)

    # Add noise
    y_noisy = y_clean + np.random.normal(0, noise_std, y_clean.shape)
    return y_noisy


# ---------------------------------------------------------------
# 3) Fixed-step Euler integrator
# ---------------------------------------------------------------
def euler_integration(fun, t0, tf, y0, n_points, *args):
    """
    Integrate y'(t) = fun(t, y, *args) from t0 to tf
    using Euler's method with exactly n_points in time.
    (So there are n_points-1 "steps".)

    :param fun:     function fun(t, y, *args) -> [dy1/dt, dy2/dt]
    :param t0:      initial time
    :param tf:      final time
    :param y0:      initial condition (2D in this example)
    :param n_points: total number of points in the Euler grid
    :param args:    extra parameters passed to fun
    :return: (t_sol, y_sol)
             t_sol: np.array of shape (n_points,)
             y_sol: np.array of shape (n_points, 2)
    """
    dt = (tf - t0) / (n_points - 1)  # uniform step
    t_sol = np.linspace(t0, tf, n_points)
    y_sol = np.zeros((n_points, len(y0)))
    y_sol[0, :] = y0

    for i in range(n_points - 1):
        # Evaluate derivative at current step
        dydt = fun(t_sol[i], y_sol[i], *args)
        # Euler update:
        y_sol[i+1] = y_sol[i] + dt * np.array(dydt)

    return t_sol, y_sol


def backward_euler_integration(fun,fun_jac, t0, tf, y0, n_points, *args,
                               max_iter=20, tol=1e-8):
    """
    Backward (implicit) Euler method for y'(t) = fun(t, y, *args).
    
    We partition [t0, tf] into n_points times, so step size is dt = (tf - t0)/(n_points-1).
    For each step:
        y_{n+1} = y_n + dt * f(t_{n+1}, y_{n+1}),
    we solve implicitly using Newton's method.
    
    :param fun:     function fun(t, y, *args) -> [dy1/dt, dy2/dt]
    :param t0:      float, initial time
    :param tf:      float, final time
    :param y0:      initial condition array-like (2D for Lotka-Volterra)
    :param n_points: total points in time grid (n_points - 1 steps)
    :param args:    extra parameters to pass into fun
    :param max_iter: maximum Newton iterations per step
    :param tol:     tolerance for Newton convergence
    :return: (t_sol, y_sol)
             t_sol: shape (n_points,)
             y_sol: shape (n_points, 2)
    """
    dt = (tf - t0) / (n_points - 1)
    t_sol = np.linspace(t0, tf, n_points)
    dim = len(y0)
    y_sol = np.zeros((n_points, dim))
    y_sol[0] = y0
    
    for n in range(n_points - 1):
        t_current = t_sol[n]
        t_next = t_sol[n+1]
        
        # We'll do a Newton iteration for y_{n+1}.
        # Start from forward-Euler guess (or simply y_sol[n])
        y_old = y_sol[n].copy()
        y_new = y_old + dt * np.array(fun(t_next, y_old, *args))  # initial guess

        for iteration in range(max_iter):
            f_val = np.array(fun(t_next, y_new, *args))  # shape (2,)
            
            # G(y_new) = y_new - y_old - dt*f_val = 0
            G = y_new - y_sol[n] - dt * f_val
            
            # If using the Lotka-Volterra Jacobian, we can do:
            # J = dG/dy_new = I - dt*(df/dy)
            # But let's call a separate function or do partial derivatives:
            J_f = fun_jac(t_next, y_new, *args)  # shape (2,2)
            
            # For more generic 'fun', you might approximate jac with finite diffs
            # but here we have an analytical expression:
            # J_G = I - dt * J_f
            I2 = np.eye(dim)
            J_G = I2 - dt * J_f  # shape (2,2)
            
            # Solve linear system J_G * delta = -G for the Newton update
            try:
                delta_y = np.linalg.solve(J_G, -G)
            except np.linalg.LinAlgError:
                # If singular, break or do something else
                print("Newton iteration: Jacobian is singular, step cannot converge!")
                break

            y_new = y_new + delta_y
            
            # Check convergence
            if np.linalg.norm(delta_y) < tol:
                break
        
        y_sol[n+1] = y_new

    return t_sol, y_sol

# ---------------------------------------------------------------
# 4) Cost function that picks the nearest Euler index
#    for each data time (no interpolation)
# ---------------------------------------------------------------
def cost_function(params, t_data, data, y0, n_euler_points):
    """
    Compute sum of squared errors between the Euler solution
    and the observed data at t_data.

    We create an Euler time grid with n_euler_points >= len(t_data),
    then for each t_data[i], we pick the nearest time index in that
    Euler grid. This avoids direct interpolation while still comparing
    all data points.

    :param params:         [alpha, beta, gamma, delta]
    :param t_data:         array of shape (n_data,) e.g. 20 points
    :param data:           array of shape (n_data, 2)
    :param y0:             initial condition
    :param n_euler_points: total Euler points from t=0 to t=t_data[-1]
    :return: scalar sum of squared residuals
    """
    alpha_est, beta_est, gamma_est, delta_est = params

    # Integrate with Euler
    t0, tf = t_data[0], t_data[-1]
    # t_euler, y_euler = euler_integration(
    #     lotka_volterra,
    #     t0,
    #     tf,
    #     y0,
    #     n_euler_points,
    #     alpha_est, beta_est, gamma_est, delta_est)

    t_euler, y_euler = backward_euler_integration(
    lotka_volterra, #fun=
    jac_lotka_volterra, # fun_jac=
    t0,
    tf,
    y0,
    n_euler_points, #n_points
    alpha_est, beta_est, gamma_est, delta_est)

    # For each data time, pick the nearest index in t_euler.
    # Because t_data and t_euler might not align exactly (except for n_euler_points=20),
    # we do a simple nearest-neighbor approach.
    residuals = []
    for i, t_i in enumerate(t_data):
        # Find nearest index
        idx = np.argmin(np.abs(t_euler - t_i))
        # Compare the Euler solution at that index to the data
        diff = y_euler[idx] - data[i]
        residuals.append(diff)

    residuals = np.array(residuals)  # shape (n_data, 2)
    return np.sum(residuals**2)


```

```python
# ---------------------------------------------------------------
# 5) Main demonstration
# ---------------------------------------------------------------
np.random.seed(0)  # for reproducible noise
noise = 0
# True parameters
alpha_true = 1.0
beta_true  = 0.5
gamma_true = 0.8
delta_true = 0.3

# Data: 20 points from t=0..19 (for instance)
n_data = 50
t_data = np.linspace(0, 19, n_data)  # from 0 to 19 (inclusive)

# Initial conditions
y0 = [1.0, 0.5]

# ---------------------------------------------------------------
# Generate synthetic "true" data
# (Using solve_ivp with small tolerances as a 'standard solver')
# ---------------------------------------------------------------
data_noisy = generate_data(alpha_true, beta_true, gamma_true, delta_true,
                           t_data, y0, noise_std=noise)

# We'll try euler points in {20, 40, 60, 80, 100}.
# Notice that only for n_euler=20 do we get an EXACT match of time grids.
# For the others, we rely on nearest-neighbor matching.
# euler_points_list = [20, 40, 60, 80, 100]
euler_points_list = [200*j for j in range(1,10)]
# euler_points_list = [20, 40, 60, 80]


fitted_params = []
for n_euler in euler_points_list:
    # Initial guess
    # initial_guess = [1.2, 0.4, 0.7, 0.25]
    initial_guess = np.random.normal(3,1, size=(4,))

    # Optimize using 'minimize' (BFGS)
    res = minimize(
        cost_function,
        x0=initial_guess,
        args=(t_data, data_noisy, y0, n_euler),
        method='BFGS'
    )

    fitted_params.append(res.x)

```

```python
# Print results
print("True parameters:")
print(f"  alpha={alpha_true:.3f}, beta={beta_true:.3f}, "
      f"gamma={gamma_true:.3f}, delta={delta_true:.3f}\n")

print("n_euler_points -> (alpha_est, beta_est, gamma_est, delta_est)")
for n_euler, (a_est, b_est, g_est, d_est) in zip(euler_points_list, fitted_params):
    print(f"{n_euler:3d} -> ({a_est:.3f}, {b_est:.3f}, {g_est:.3f}, {d_est:.3f})")

# ---------------------------------------------------------------
# 6) (Optional) Plot the final fit for each scenario
# ---------------------------------------------------------------
fig, axes = plt.subplots(1, len(euler_points_list), figsize=(18, 4), sharey=True)

for ax, n_euler, (a_est, b_est, g_est, d_est) in zip(axes, euler_points_list, fitted_params):
    # Solve once more with Euler and these fitted params
    # Data: 20 points from t=0..19 (for instance)
    try:
        n_data_sim = 100
        t_sim = np.linspace(0, 19, n_data_sim)  # from 0 to 19 (inclusive)
        
        # ---------------------------------------------------------------
        # Generate synthetic "true" data
        # (Using solve_ivp with small tolerances as a 'standard solver')
        # ---------------------------------------------------------------
        print("readhed her")
        y_sim = generate_data(a_est, b_est, g_est, d_est,
                           t_sim, y0, noise_std=0)
        # t_euler, y_euler = euler_integration(
        #     lotka_volterra, t_data[0], t_data[-1], y0, n_euler,
        #     a_est, b_est, g_est, d_est
        # )
    
        # Plot data
        ax.plot(t_data, data_noisy[:, 0], 'o', label='Data x', alpha=0.5)
        ax.plot(t_data, data_noisy[:, 1], 'o', label='Data z', alpha=0.5)
    
        # Plot Euler solution
        ax.plot(t_sim, y_sim[:, 0], '-', label='Euler x')
        ax.plot(t_sim, y_sim[:, 1], '-', label='Euler z')
    
        ax.set_title(f'Euler points = {n_euler}')
        ax.set_xlabel('Time')
    except:
        None

axes[0].set_ylabel('Population')
axes[0].legend(loc='best', fontsize='small')
plt.tight_layout()
plt.show()
```
